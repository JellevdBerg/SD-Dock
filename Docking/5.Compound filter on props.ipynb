```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
data = pd.read_csv('./Enamine_predictions_with_stats.csv')
```

```python
# Generate histograms for MW, NumRotatableBonds, and NumRings
data[['MW', 'NumRotatableBonds', 'NumRings']].hist(bins=10, figsize=(10, 5))
plt.savefig("./Results/PNGs/Property_statistics_histo.png", dpi=300)
plt.show()
```

```python
# filter out the compounds with 350<MW<450, NumRotatableBonds =<7, and 3<NumRings<5
filtered_data = data[(data['MW'] > 350) & (data['MW'] < 450) & (data['NumRotatableBonds'] <= 7) & (data['NumRings'] > 3) & (data['NumRings'] < 5)]

# save the filtered data to a new csv file
filtered_data.to_csv('./filtered_data.csv', index=False)
```
